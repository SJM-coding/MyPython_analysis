```python
import requests
import os
from bs4 import BeautifulSoup

url = 'https://comic.naver.com/webtoon/detail?titleId=833255&no=35&week=tue'
#웹툰에 따라 titleID가 달라지고 no에서는 회차를 표현함.
def download_one_episode(title,no,url):
    req_header = {
        'referer' : url
    }

    res = requests.get(url)

    if res.ok:
        
        soup = BeautifulSoup(res.text, 'html.parser')
        print(len(soup.select("img[src*= 'IMAG01']"))) # img [src = 'IMAG01'] -> 이렇게 쓰면 imgurl이 IMAG01과 아예 같아야함
        print(len(soup.select("img[src$='.jpg']")))
        print(len(soup.select("img[src^='https://image-comic']")))
        
        img_url_list = list() 
        img_tags = soup.select("img[src*= 'IMAG01']")
        for img_tag in img_tags:
            img_src = img_tag['src']
            img_url_list.append(img_src)
        
        print(len(img_url_list),img_url_list[0:3])
        
        #img 디렉토리가 없으면 생성
        imgdir_name = 'img'
        if not os.path.isdir(imgdir_name):
            os.mkdir(imgdir_name)
            
        title_dir = os.path.join(imgdir_name, title)
        if not os.path.isdir(title_dir):
            os.mkdir(title_dir)

        episode_dir = os.path.join(title_dir, str(no))
        if not os.path.isdir(episode_dir):
            os.mkdir(episode_dir)
        for img_url in img_url_list:

            if res.ok:
                img_data = res.content
                file_name = os.path.basename(img_url)
                path = os.path.join(episode_dir,file_name)
                print(file_name)
                # binday data를 file에 write하기
                with open(path,'wb') as file:
                    print(f'writing to {path}({len(img_data):,} bytes)')
                    file.write(img_data)    

    else:
        print(f'Error Code = {res.status.code}')
        
        
download_one_episode('낢이사는이야기',48,'https://comic.naver.com/webtoon/detail?titleId=833255&no=49&week=tue')
```
