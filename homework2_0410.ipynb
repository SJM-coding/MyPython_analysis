```python
import requests
import os
from dotenv import load_dotenv
import pandas as pd
import re
import json

# .env 파일에서 환경 변수 로드
load_dotenv()

def search_shop(key='한글'):
# 환경 변수에서 값 읽기
    client_id = os.getenv("NAVER_CLIENT_ID")
    client_secret = os.getenv("NAVER_CLIENT_SECRET")

    headers = {
        'X-Naver-Client-Id': client_id,
        'X-Naver-Client-Secret': client_secret,
    }

    payload = {
        'query': key,
        'display': 50,
        'sort': 'date'
    }

    url = 'https://openapi.naver.com/v1/search/shop.json'

    # 요청 보내기
    res = requests.get(url, params=payload, headers=headers)
    items_data = res.json()['items']

    # 결과 저장 리스트
    items_list = list()
    item_list = []


    for i, item in enumerate(items_data):
        title = re.sub(r'<.*?>', '', item['title'])

        item_list.append(title)
        item_list.append(item['link'])
        item_list.append(item['image'])
        item_list.append(item['lprice'])
        item_list.append(item['hprice'])
        item_list.append(item['productId'])
        item_list.append(item['productType'])
        item_list.append(item['brand'])
        item_list.append(item['maker'])
        item_list.append(item['category1'])
        item_list.append(item['category2'])
        item_list.append(item['category3'])
        item_list.append(item['category4'])        
        items_list.append(item_list)
        item_list = [] 

    df = pd.DataFrame(
        items_list,
        columns=["title", "link", "image", "lprice", "hprice", "productId", "productType", "brand", "maker","category1","category2","category3","category4"]
    )

    df.to_json('data/shop.json', orient='records', force_ascii=False, indent=2)

    # 불러오기
    shop_df = pd.read_json('data/shop.json')
        # 출력

    return shop_df


df = search_shop('가디건')
print(df)
```
